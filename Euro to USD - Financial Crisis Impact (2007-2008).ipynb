### Moving Windows

* 7 days
* 30 Days
* 50 Days
* 100 Days
* 365 Days

```python
euro_dollar.USD.rolling(7).mean().head(50)
```

```python
rolling = [7, 30, 50, 100, 365]
plots = [2, 3, 4, 5, 6]
```

```python
plt.figure(figsize = (11, 9))
plt.subplot(3, 2, 1)
plt.plot(euro_dollar["Date"], euro_dollar.USD)
plt.title("Original")

for rol, plot in zip(rolling, plots):
    plt.subplot(3, 2, plot)
    plt.plot(euro_dollar["Date"], euro_dollar.USD.rolling(rol).mean(),)
    plt.title(f"Moving Average {rol}")

plt.show()    
```

```python
plt.figure(figsize = (11, 9))
plt.subplot(3, 2, 1)
plt.plot(euro_dollar["Date"], euro_dollar.USD)
plt.title("Original")

for rol, plot in zip(rolling, plots):
    plt.subplot(3, 2, plot)
    plt.plot(euro_dollar["Date"], euro_dollar.USD.rolling(rol).mean(), linewidth = 3)
    plt.title(f"Moving Average {rol}")
    
plt.tight_layout()    

plt.show()    
```

```python
euro_dollar["Rolling_mean_30"] = euro_dollar.USD.rolling(30).mean()
```

# During Financial Crisis

```python
bol_2006_2009 = (euro_dollar.Date.dt.year >= 2006) & (euro_dollar.Date.dt.year <= 2009)
```

```python
bol_2007_2008 = (euro_dollar.Date.dt.year >= 2007) & (euro_dollar.Date.dt.year <= 2008)
```
