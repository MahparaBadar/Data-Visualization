# items()

* method of dictionary which returns keys and values

```python
for day, df in traffic_per_day.items():
    plt.plot(df["Hour (Coded)"], df["Slowness in traffic (%)"])
    plt.title(day)
    plt.show()
```

# xlim([ ]), ylim([ ])

```python
for day, df in traffic_per_day.items():
    plt.plot(df["Hour (Coded)"], df["Slowness in traffic (%)"])
    plt.title(day)
    plt.ylim([0, 25])
    plt.show()
```

# All days in one plot through matplotlib.pyplot

```python
for day, df in traffic_per_day.items():
    plt.plot(df["Hour (Coded)"], df["Slowness in traffic (%)"], label = day)

plt.legend()    
plt.ylim([0, 25])
plt.show()
```
