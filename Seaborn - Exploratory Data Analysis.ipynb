### relplot

In Seaborn, a scatter plot is referred to as a "relational plot." This term highlights that this type of plot is specifically designed to visualize and explore the relationships or associations between different variables in your data, making it a powerful tool for analyzing patterns and connections within your dataset.

### kdeplot

Seaborn's kdeplot function allows you to create a smooth, histogram-like plot, where bars are replaced with a continuous curve. This curve represents the Kernel Density Estimation (KDE) of the data distribution, providing a more nuanced and continuous view of the data compared to traditional histograms, which use discrete bars to represent data points.

## sizes

```python
sns.relplot(data = pro, x = 'Gr Liv Area', y = 'SalePrice',
           hue = "Overall Cond", palette = "RdYlBu",
           size = "Garage Area", sizes = (1, 300))

plt.show()
```

```python
sns.relplot(data = pro, x = 'Gr Liv Area', y = 'SalePrice',
           hue = "Overall Cond", palette = "RdYlBu",
           size = "Rooms")    # if we use any categorical data in size (here e.g. Rooms) then we can not pass tuple in "sizes",
                              # then we will have to give list or dictionary b/c tuple is used for the ranges.
                               # range(0, 29)

plt.show()
```

```python
sns.relplot(data = pro, x = 'Gr Liv Area', y = 'SalePrice',
           hue = "Overall Cond", palette = "RdYlBu",
           size = "Rooms", sizes =[1, 200])    # if we use any categorical data in size (here e.g. Rooms)
                                               # then we can not pass tuple in "sizes",
                                              # then we will have to give list or dictionary b/c tuple is used for the ranges.
                                              # range(0, 29)

plt.show()
```
