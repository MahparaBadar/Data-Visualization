 instant: record index
 
	- dteday : date
    - season : season (1:winter, 2:spring, 3:summer, 4:fall)
	- yr : year (0: 2011, 1:2012)
	- mnth : month ( 1 to 12)
	- hr : hour (0 to 23)
	- holiday : weather day is holiday or not
	- weekday : day of the week
	- workingday : if day is neither weekend nor holiday is 1, otherwise is 0.
    + weathersit : 
		- 1: Clear
		- 2: Mist
		- 3: Light Rain 
		- 4: Heavy Rain 
	- temp : Normalized temperature in Celsius. The values are derived via (t-t_min)/(t_max-t_min), t_min=-8, t_max=+39 (only in hourly scale)
	- atemp: Normalized feeling temperature in Celsius. The values are derived via (t-t_min)/(t_max-t_min), t_min=-16, t_max=+50 (only in hourly scale)
	- hum: Normalized humidity. The values are divided to 100 (max)
	- windspeed: Normalized wind speed. The values are divided to 67 (max)
	- casual: count of casual users
	- registered: count of registered users
	- cnt: count of total rental bikes including both casual and registered

# Group Frequency Table

When we have a large number of unique values, then Group Frequency Table is used.

````python
bike.cnt.value_counts(bins = 10)   # bins = no. of intervals
````

**In the realm of statistics, many experts advise using 10 bins. This is because if you use only 5 bins, you might miss important details, and if you use too many bins, it can get confusing. So, 10 bins are often recommended because they give you a good balance between having a comprehensive view and not overwhelming you with too much information.**

````python
bike.cnt.value_counts(bins = 10).sort_index()
````

# Differences b/w Histogram and Bar Chart

(1) In a bar chart, you find individual categorical values on the x-axis, while in a histogram, you have continuous intervals.

(2) Bar charts have gaps between the bars, whereas histograms have bars that touch each other.

````python
plt.hist(bike["casual"])
plt.show() 

# this is the positive/right skewed graph  (body-->left side, tail-->right side) 
````

We have two types of distributions. One is a **symmetric distribution**, and the other is a **skewed distribution**.

**Symmetric**: If you divide something right down the middle, both halves look like mirror images of each other. In symmetric distributions, there are two main types. One is the ```normal distribution``` (often called the bell-shaped curve), also known as the ```Gaussian distribution```. In a normal distribution, there's a peak in the center, and then it gradually decreases on both sides.

**Uniform distribution**: This is the other type of symmetric distribution, where data is spread out evenly, without a peak in the center.
