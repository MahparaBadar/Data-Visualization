```python
wnba["Experience"].value_counts().sort_index()  # R = rookie (freshies)
```

```python
wnba["Experience"] = wnba["Experience"].str.replace("R", "0").astype(int)
```

```python
wnba["Experience"].value_counts().sort_index()
```

```python
# difference, direction size 
# ratio and interval data type
```

```python
def exp_ord(exp):
    if exp == 0:
        return "Rookie"
    elif exp in range(1, 4):
        return "Little Experience"
    elif exp in [4, 5]:
        return "Experienced"
    elif exp in range(6, 11):
        return "Very Experienced"
    else:
        return "Veteran"
        
```

```python
wnba["Experience"].shape
```

```python
wnba["Experience"]
```

```python
wnba["Experience"].apply(exp_ord)
```

```python
wnba["Exp_Ord"] = wnba["Experience"].apply(exp_ord)
```

```python
wnba["Exp_Ord"].value_counts()
```

```python
wnba["Exp_Ord"].value_counts().iloc[[-2, 0, 2, 1, -1]]
```

```python
wnba["Exp_Ord"].value_counts().iloc[[-2, 0, 2, 1, -1]].plot.barh(title = "No. of Players in WNBA by Level of Experience")
plt.show()
```

```python
wnba["age_cat"] = wnba["Age"].apply(lambda age: "Old" if age >= 27 else "Young")
wnba["min_cat"] = wnba["MIN"].apply(lambda Min: "avg or above" if Min >= 497 else "below avg")
```

```python
sns.set_theme()

sns.countplot(x = "age_cat", hue = "min_cat", data = wnba)

plt.show()
```

```python
wnba.groupby("age_cat")["MIN"].plot.hist()
plt.show()
```

```python
wnba.groupby("age_cat")["MIN"].plot.hist()
wnba.groupby("age_cat")["MIN"].plot.hist()
plt.show()
```

```python
wnba[wnba["Age"] <= 27]["MIN"].plot.hist(label = "Young")
wnba[wnba["Age"] > 27]["MIN"].plot.hist(label = "Old")
plt.legend()
plt.show()
```

```python
wnba[wnba["Age"] <= 27]["MIN"].plot.hist(label = "Young", histtype = "step")
wnba[wnba["Age"] > 27]["MIN"].plot.hist(label = "Old", histtype = "step")
plt.legend()
plt.show()
```

```python
wnba[wnba["Age"] <= 27]["MIN"].plot.hist(label = "Young", histtype = "step")
wnba[wnba["Age"] > 27]["MIN"].plot.hist(label = "Old", histtype = "step")
plt.legend(loc = "lower left")   # location
plt.axvline(497, label = "Average", alpha = 0.3)
plt.show()
```
