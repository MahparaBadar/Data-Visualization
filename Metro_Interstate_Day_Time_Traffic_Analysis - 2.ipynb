```python
weekend = day[~ (day["dayofweek"] <= 4)]  # in pandas "not" is written as "~"
weekend
```

```python
# In pivot_table, "mean" is already set at default

hourly_weekdays_avg = weekdays.pivot_table(index = "hour", values = "traffic_volume")
hourly_weekdays_avg
```

```python
hourly_weekend_avg = weekend.pivot_table(index = "hour", values = "traffic_volume")
hourly_weekend_avg
```

```python
hourly_weekdays_avg.index
```

```python
plt.figure(figsize = (11, 4))

plt.subplot(1, 2, 1)

plt.plot(hourly_weekdays_avg.index, hourly_weekdays_avg)
plt.ylim(1500, 6250)

plt.subplot(1, 2, 2)

plt.plot(hourly_weekend_avg.index, hourly_weekend_avg)
plt.ylim(1500, 6250)

plt.show()
```
