# Gestalt principles

https://medium.com/geekculture/gestalt-psychology-as-applied-in-product-design-941b87b01218

Gestalt principles are a set of guidelines that describe how people tend to perceive visual elements as organized wholes, rather than isolated parts. In easy words, these principles help us understand how our brains naturally group and make sense of what we see. Here are a few key Gestalt principles:

1) **Proximity:** When you place related data points close to each other, viewers are likely to perceive them as a group. 

2) **Similarity:** Making data elements that share common characteristics look similar helps viewers associate them. This could involve using the same color, shape, or size for data points that belong to the same category.

3) **Enclosure:** Grouping data points together within a boundary, such as a box or a shape, can emphasize their collective significance. This is particularly useful when you want to highlight specific sets of data.

4) **Connectivity:** Connectivity refers to the idea that elements visually linked or connected by lines, paths, or other visual cues are perceived as a unified group. This principle underscores the importance of visual connections in shaping our perception of relationships and unity within a visual composition.

5) **Continuity:** Creating smooth and continuous lines in your data visualization helps guide the viewer's eye along a logical path. This principle can be applied to show trends or connections between data points.

6) **Figure-Ground:** Establishing a clear distinction between the main data points (figure) and the background helps viewers focus on the essential information. For instance, emphasizing the key data points while keeping the background less prominent.

# Preattentive Attributes

https://daydreamingnumbers.com/blog/preattentive-attributes-example/

Preattentive attributes refer to visual properties that our brains automatically and rapidly process without conscious effort. These attributes can quickly grab our attention because our brains are wired to detect them effortlessly. Here are some common preattentive attributes:

1) **Color:** Bright or contrasting colors can quickly attract attention to specific data points.

2) **Size:** Varied sizes of elements help emphasize importance within a visual representation.

3) **Orientation:** Distinct orientations, such as differently angled bars in a chart, can stand out effortlessly.

4) **Shape:** Unique shapes or deviations from regular shapes draw immediate notice in a visualization.

5) **Motion:** Animated or moving elements can effectively capture and direct viewer attention.

```python
rolling = [7, 30, 50, 100, 365]

plots = [2, 3, 4, 5, 6]
```

```python
plt.figure(figsize = (9, 11))

plt.subplot(3, 2, 1)
plt.plot(euro_to_dollar["Date"], euro_to_dollar.USD)
plt.title("Original")

for rol, plot in zip(rolling, plots):
    plt.subplot(3, 2, plot)
    plt.plot(euro_to_dollar["Date"], euro_to_dollar.USD.rolling(rol).mean())
    plt.title(f"Moving Average {rol} (Days)")
    
plt.tight_layout()
    
plt.show()
```

```python
plt.figure(figsize = (9, 11))

plt.subplot(3, 2, 1)
plt.plot(euro_to_dollar["Date"], euro_to_dollar.USD)
plt.title("Original")

for rol, plot in zip(rolling, plots):
    plt.subplot(3, 2, plot)
    plt.plot(euro_to_dollar["Date"], euro_to_dollar.USD.rolling(rol).mean(), linewidth = 3)
    plt.title(f"Moving Average {rol} (Days)")
    
plt.tight_layout()
    
plt.show()
```
