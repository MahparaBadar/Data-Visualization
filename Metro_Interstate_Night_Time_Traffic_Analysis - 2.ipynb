```python
plt.figure(figsize = (11, 4))

plt.subplot(1, 2, 1)

hourly_weekday_avg = weekdays.groupby("hour")["traffic_volume"].mean()
plt.plot(hourly_weekday_avg.index, hourly_weekday_avg)


plt.subplot(1, 2, 2)

hourly_weekend_avg = weekend.groupby("hour")["traffic_volume"].mean()
plt.plot(hourly_weekend_avg.index, hourly_weekend_avg)


plt.show()
```

```python
plt.figure(figsize = (11, 4))

plt.subplot(1, 2, 1)

hourly_weekday_avg = weekdays.groupby("hour")["traffic_volume"].mean()
plt.plot(hourly_weekday_avg.index, hourly_weekday_avg)
plt.title("Weekdays")
plt.xlabel("Hour")
plt.ylabel("Traffic Volume")
plt.ylim(0, 6000)


plt.subplot(1, 2, 2)

hourly_weekend_avg = weekend.groupby("hour")["traffic_volume"].mean()
plt.plot(hourly_weekend_avg.index, hourly_weekend_avg)
plt.title("Weekend")
plt.xlabel("Hour")
plt.ylim(0, 6000)

plt.show()
```

```python
hourly_weekday_avg = weekdays.groupby("hour")["traffic_volume"].mean()
plt.plot(hourly_weekday_avg.index, hourly_weekday_avg, label = "Weekdays")

hourly_weekend_avg = weekend.groupby("hour")["traffic_volume"].mean()
plt.plot(hourly_weekend_avg.index, hourly_weekend_avg, label = "Weekend")

plt.legend()
plt.xlabel("Hour")
plt.ylabel("Traffic Volume")

plt.show()
```

```python
plt.figure(figsize = (11,4))
plt.subplot(1,2,1)

hourly_weekday_avg = weekdays.pivot_table(index = "hour", values="traffic_volume")
plt.plot(hourly_weekday_avg.index, hourly_weekday_avg)
plt.ylim(0,6500)

plt.subplot(1,2,2)

hourly_weekend_avg = weekend.pivot_table(index = "hour", values="traffic_volume")
plt.plot(hourly_weekend_avg.index, hourly_weekend_avg)
plt.ylim(0,6500)

plt.show()
```
