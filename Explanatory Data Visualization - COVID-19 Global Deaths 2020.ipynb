# Familiarity Principle

Familiarity Principle refers to the idea that data visualizations should leverage familiar symbols, layouts, and conventions that the audience is accustomed to. This helps viewers quickly understand and interpret the information presented, as they can rely on their prior knowledge and experiences with similar visualizations.

# Design principles

Design principles are fundamental guidelines that help you create visually appealing and effective designs. They include ideas like balance, contrast, alignment, and unity, which help you organize and present information in a way that's both visually pleasing and easy to understand.

There are two interfaces of matplotlib

#### Functional Approach / Interface

* (plt. plot, plt.scatter, plt.bar, plt.title, plt.show etc)
* provides less power for customization
* exploratory data analysis

#### Object Oriented Interface (OO Interface)

* provides customization
* explanatory data analysis

**Matplotlib Interfaces:**

1. **Pyplot Interface**: Quick and easy way for basic plots, similar to MATLAB's functionality. Use functions like `plot()`, `scatter()`, and `hist()` for fast visualizations.

2. **Object-Oriented Interface**: Offers more control and flexibility. Create Figure and Axes objects for detailed manipulation, especially useful for complex or multi-panel plots.

3. **Pyplot-style and Object-Oriented Combined**: We can use both interfaces together, benefiting from simplicity and customization power.

These interfaces enable us to create various data visualizations, such as line charts, bar charts, scatter plots, histograms, and more, while tailoring them to your specific data visualization needs.

Design principles help us in two ways, first of all they generate a design options according to data like in this case we can opt bar and pie charts. Secondly it helps us in chosing the best option from those options.

Design principles serve a dual purpose:

1. **Generating Design Options**: They provide a framework to create various design choices based on the data. For instance, we can consider options like bar charts and pie charts when visualizing data.

2. **Selecting the Best Option**: Design principles also guide you in the process of choosing the most suitable design option from the available choices. This ensures that our data visualization effectively communicates the information we want to convey. This should be elaborated fully on the familiarity principle.

There are three elemnts of graph:

1. **Data Element:**
   The data element is a fundamental part of your graph. It represents the individual data points that you are plotting. These data elements have labels associated with them, which provide information about the data they represent. In the context of a graph, data elements are typically plotted along the x-axis, showing how your data varies.
   

2. **Structure Element:**
   The structure element refers to the graphical components that define the overall structure and layout of your graph. This includes elements such as the axes (both x and y), lines connecting data points, the point where the axes intersect (the origin), and the small marks or ticks on the axes. Sometimes, you may use a legend to explain the meaning of different colors or symbols, and all of these components together are considered part of the structure element.
   

3. **Decorations:**
   Decorations encompass various visual enhancements and artistic elements that you add to your graph to make it more appealing or informative. This can include different shapes, colors, or drawings that are not directly related to the data but are added for aesthetic or explanatory purposes. In other words, decorations are the artistic elements you use to enhance the visual appeal and understanding of your graph.
